```python
# public,private,protected

# public mode
class emp:
    def abc(self,a,b,c):
        c = a+b+c
        return c
    
e = emp()
e.abc(1,2,3)
        
```

```python
# private mode
class emp:
    
    
        
    def __abc(self,a,b,c):
        c = a+b+c
        return c
    
    
    
e = emp()
e.__abc(1,2,3)
```

```python
# protected mode
class emp:
    def _abc(self,a,b,c):
        c = a+b+c
        return c
    
e = emp()
e._abc(1,2,3)
```

# Abstarction

```python
from abc import ABC,abstractmethod
class emp(ABC):
    @abstractmethod
    def info(self):
        None
        
class emp2(emp):
    def info(self):
        print("Hello bhai!!")
        
e = emp2()
e.info()
```

```python
from abc import ABC,abstractmethod
class first(ABC):
    @abstractmethod
    def info(self):
        print("first class")
    
class sec(first):
    def details(self):
        print("second class op")
        
s = sec()
s.details()
```
