```python
# A shop will give discount of 10% if the cost of purchased 
# quantity is more than 1000.
# Ask user for quantity
# Suppose, one unit will cost 100.
# Judge and print total cost for user.

unit_cost = int(input("Enter the unit cost"))
unit_purchase = int(input("Enter the unit purchase"))
price = unit_cost * unit_purchase

if price > 1000:
    discount = (price*10)/100
    excat_amount_paid = price - discount
    print("Your discount is ----->",discount)
    print("Your excat price to be paid is ----->",excat_amount_paid)
else:
    print("total price is ---->",price)
    print("No discount")

```

```python
# A student will not be allowed to sit in exam 
# if his/her attendance is less than 75%.
# Take following input from user
# Number of classes held
# Number of classes attended.
# And print
# percentage of class attended
# Is student is allowed to sit in exam or not.

num_of_class_held = int(input("Enter the class held"))
num_of_class_attend = int(input("Number of class attend"))

per_of_class_attend = (num_of_class_attend*100)/num_of_class_held

if per_of_class_attend<75:
    medical_condition = input("Do you have any medical condition for yes type Y and for no type N")
    if medical_condition == 'Y':
        print("Allowed due to medical condition")
    else:
        print("Not Allowed")
    
else:
    print("Allowed")
print(per_of_class_attend)
```
