```python
a=[78,90,6,54,33,12,86]
greatest_num=a[0]
for i in a:
    if i>greatest_num:
        greatest_num=i
print(greatest_num,"is the first greatest number")
second_greatest=a[0]
for j in a:
    if j>second_greatest and j<greatest_num:
        second_greatest=j
print(second_greatest,"is the second greatest number")
```
