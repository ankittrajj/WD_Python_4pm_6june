# Filter

```python
#wap to take the data from the list where age is less than 30.

a = [45,56,12,34,89,90,34,54,88,23]
c = list(filter(lambda a:True if a<30 else False,a))
```

```python
print(c)
```

```python
a = [45,56,12,34,89,90,34,54,88,23]
def fun(a):
    if a<30:
        return True
    else:
        return False
    
list(filter(fun,a))
```
