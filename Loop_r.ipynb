```python
#wap to find prime number.

num = int(input("Enter the number"))
if num>1:
    for i in range(2,num):
        if num % i == 0 :
            print(num,"Not the prime number")
            break
            
        else:
            print(num,"Prime number")
else:
    print(num,"not prime number")
            
```
