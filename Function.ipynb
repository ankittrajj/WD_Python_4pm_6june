```python
# types of user defined function.
# 1. no return no arg-------> worst
# 2. no return with arg
# 3. with return with arg-----> best
# 4. with return no arg
```

```python
# no return no arg
def add():
    a = 8
    b = 2
    c = a+b
    print(c)
    
add()
```

```python
# no return with arg
def add(x,y):
    z = x+y
    print(z)
    
add(1,2)
```

```python
# with return with arg
def mul(x,y):
    z = x*y
    return z 
    

    
mul(3,4)
```

```python
def max_num(a,b,c):
    if b>a:
        if c>b:
            return c
        else: return b
    elif c>a:
        return c
    else: return a
    
print(max_num(7,31,20))
```
