```python
class parent:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    
    
class child(parent):
    def __init__(self,name,age,salary):
        parent.__init__(self,name,age)
        self.salary= salary
        
        
c = child("JOJO",23,12)
```

```python
c.salary
```

```python
c.name
```

```python
# gpc to pc
class gpc:
    def __init__(self,name,age):
        self.name = name
        self.age = age
        
    
    
class pc(gpc):
    def __init__(self,name,age,salary):
        super().__init__(name,age)
        self.salary = salary
    

class cc:
    def __init__(self,roll_num):
        self.roll_num = roll_num
        
        
p = pc("A",34,45)
```

```python
# gpc to pc to child

class gpc:
    def __init__(self,name,age):
        self.name = name
        self.age = age
        
    
    
class pc(gpc):
    def __init__(self,name,age,salary):
        super().__init__(name,age)
        self.salary = salary
    

class cc(pc):
    def __init__(self,name,age,salary,roll_num):
        super().__init__(name,age,salary)
        self.roll_num = roll_num
        
        
c = cc("Yash",12,34,1)
```

```python
c.salary
```

```python
c.roll_num
```

```python
c.name
```

# Encapsulation

```python
# public mode
class me:
    a = 10
    print(a)
me()
```

```python
# private mode
class me:
    __a = 10
    print(a)
me()
```

```python
# protected mode
class me:
    _a = 10
    print(_a)
me()
```
