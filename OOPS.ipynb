```python
#wap to find out the speed of a train which covers a distance
# of 40km in 2 hours?(hint---> use class and object)


class train:
    def speed_of_train(self,dis,time):
        speed = dis/time
        print(speed)
    
t = train()
t.speed_of_train(40,2)
```

```python
#wap to find the avg weather of a place which is:
# 32,34,23,12,40
class climate:
    def info(self,a,b,c,d,e):
        avg = (a+b+c+d+e)/5
        return avg
    
c = climate()
c.info(32,34,23,12,40)
```
